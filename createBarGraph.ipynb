```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
fileName = "./biCsvs/htmlBarGraph.csv"
```

```python
def importData(filename):
    X = []
    y = []
    with open(filename, 'r') as training_data:
        while True:
            line = training_data.readline().rstrip()
            if line == "":
                break
            data = [ float(x) for x in line.split(',')[:-1] ]
            X.append(data[:])
            y.append(line.split(',')[-1])
    return X, y
```

```python
yAxis, xAxis = importData(fileName)
```

```python
len(yAxis[0])
```

```python
N = 2
for i in range(len(xAxis)):
    newArray = []
    for j in range(256):
        start = j*256
        end = start+255
        arr = yAxis[i][start:end]
        maxElements = sorted(arr, reverse = True)[:N]
        newArray += maxElements
    yAxis[i] = newArray
```

```python
xAxis = np.arange(start=0, stop=256*N)
```
